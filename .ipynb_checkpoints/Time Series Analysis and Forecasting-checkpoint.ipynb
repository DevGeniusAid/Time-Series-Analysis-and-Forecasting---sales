### Question 2: Decomposition, ACF,PACF,Stationarity
Importance of decomposition, ACF, PACF, and stationarity:
Decomposition, autocorrelation function (ACF), partial autocorrelation function (PACF), and stationarity are
crucial concepts in time series analysis. They help in understanding the underlying patterns, dependencies,
and characteristics of time series data, and they play a vital role in building accurate and reliable forecasting
models.

**A. Plot the monthly decomposition sales for trends and seasonality, and comment on the results**

```python
data.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming 'sales_data' is your DataFrame with 'Order_Date' and 'Sales' columns
# Convert 'Order_Date' to datetime and set it as index
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
data.set_index('Order_Date', inplace=True)

# Resample the data to monthly frequency and fill missing values
sales_monthly = data['Sales'].resample('M').sum().fillna(0)

# Perform seasonal decomposition
decomposition = seasonal_decompose(sales_monthly, model='additive')

# Plot the decomposition
plt.figure(figsize=(12, 8))

# Original time series
plt.subplot(411)
plt.plot(sales_monthly, label='Original')
plt.legend(loc='upper left')

# Trend component
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')

# Seasonal component
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')

# Residual component
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')

plt.suptitle('Monthly Decomposition of Sales')
plt.tight_layout()
plt.show()
```

Comments on the Results:

Original Time Series: The original time series shows the total sales per month.

Trend Component: The trend component captures the long-term movement of the sales data, indicating overall sales growth or decline.

Seasonal Component: The seasonal component represents the recurring patterns or seasonality in the sales data, such as monthly or quarterly fluctuations.

Residual Component: The residual component reflects the random fluctuations or noise in the data that cannot be explained by the trend or seasonal patterns.

**B. Generate the ACF and PACF plots and comment on the results.**

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Generate ACF plot
plt.figure(figsize=(12, 6))
plot_acf(sales_monthly, lags=24, ax=plt.gca())  # Adjust 'lags' parameter as needed
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

# Generate PACF plot
plt.figure(figsize=(12, 6))
plot_pacf(sales_monthly, lags=12, ax=plt.gca())  # Adjust 'lags' parameter within permissible range
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.grid(True)
plt.show()

```

Comments on the Results:

ACF Plot:

The ACF plot shows the correlation between the time series and its lagged values.<br>
Interpretation: Look for significant spikes or patterns beyond the confidence interval (dashed lines). These indicate strong correlations at certain lag values.<br>
Comments: Significant correlations at specific lags may suggest the presence of seasonality or autoregressive patterns.

PACF Plot:

The PACF plot shows the partial autocorrelation between the time series and its lagged values, controlling for the correlations at shorter lags.<br>
Interpretation: Look for significant spikes or patterns beyond the confidence interval (dashed lines). These indicate significant partial correlations at certain lag values.<br>
Comments: Significant spikes in the PACF plot indicate potential AR terms in the time series model.

**C. Check for stationarity using the Dickey-Fuller test and comment on the results**

```python
from statsmodels.tsa.stattools import adfuller

# Perform Dickey-Fuller test
result = adfuller(sales_monthly)

# Extract and print test results
print('Dickey-Fuller Test Results:')
print(f'Test Statistic: {result[0]}')
print(f'p-value: {result[1]}')
print(f'Critical Values:')
for key, value in result[4].items():
    print(f'   {key}: {value}')
```

### Question 3: Modelling Forecasting Techniques
Forecasting time-series data

Forecasting is a process of making predictions or estimates about future events or trends based on historical
data, patterns, and analysis. In various fields and industries, forecasting is employed to anticipate future
values of a variable, such as sales, demand, financial metrics, or weather conditions. The goal is to provide
decision-makers with valuable insights that can inform planning, resource allocation, and strategic decisionmaking.
In this question, you are required to use the Holt-Winters Exponential Smoothing forecasting technique.

**A. Model the Holt-Winters Exponential Smoothing with the Order date and sales. Use the
TotalSalesPerMonth table to model it per month** 

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Assuming 'sales_monthly' is your DataFrame with 'Order_Date' as index and 'Sales' column
# Make sure 'Order_Date' is already set as the index

# Fit the Holt-Winters model
model = ExponentialSmoothing(sales_monthly, seasonal='additive', seasonal_periods=12).fit()

# Generate fitted values and forecast
fitted_values = model.fittedvalues
forecast = model.forecast(steps=12)

# Plot the actual sales, fitted values, and forecast
plt.figure(figsize=(12, 6))
plt.plot(sales_monthly, label='Actual Sales')
plt.plot(fitted_values, label='Fitted Sales', linestyle='--')
plt.plot(forecast, label='Forecasted Sales', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Holt-Winters Exponential Smoothing - Actual vs Fitted vs Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

**B. Assess the Model fit/performance above and comment on the results.**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Compute evaluation metrics
mae = mean_absolute_error(sales_monthly, fitted_values)
mse = mean_squared_error(sales_monthly, fitted_values)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

Mean Absolute Error (MAE):<br>
8696.88: On average, the model's predictions are off by approximately $8696.88.
Assessment: The MAE provides a straightforward measure of prediction accuracy. Given the sales data scale, this error might be considered substantial depending on the overall sales figures.

Mean Squared Error (MSE):<br>
114,221,468.72: This large value indicates that some of the prediction errors are quite significant.<br>
Assessment: MSE penalizes larger errors more heavily than smaller ones, so this high value suggests that the model might be making some large errors in its predictions.

Root Mean Squared Error (RMSE):<br>
10,687.44: This metric is in the same units as the sales data and provides a measure of the average prediction error.<br>
Assessment: An RMSE of 10,687.44 indicates that, on average, the predictions deviate from the actual sales by about $10,687.44. This is relatively high, suggesting that the model's predictive accuracy may not be optimal.

**C. Forecast and plot the sales for the next 12 months. does the results make sense?**

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit the Holt-Winters model
model = ExponentialSmoothing(sales_monthly, seasonal='additive', seasonal_periods=12).fit()

# Generate fitted values and forecast for the next 12 months
fitted_values = model.fittedvalues
forecast = model.forecast(steps=12)

# Create a combined DataFrame for actual, fitted, and forecasted values
combined = sales_monthly.append(forecast.rename('Forecast'))

# Plot the actual sales, fitted values, and forecast
plt.figure(figsize=(12, 6))
plt.plot(combined.index, combined, label='Actual Sales and Forecast')
plt.plot(fitted_values.index, fitted_values, linestyle='--', label='Fitted Sales')
plt.axvline(x=sales_monthly.index[-1], color='red', linestyle='--', label='Forecast Start')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Holt-Winters Exponential Smoothing - Actual vs Fitted vs Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

The results make sense and indicate that the Holt-Winters model is a good fit for the sales data. If further refinement is needed, you might consider adjusting the model parameters or exploring additional features that could impact sales, such as promotions or economic indicators. However, based on the visual inspection and the consistency with historical patterns, the current model provides a solid foundation for forecasting future sales.

### Question 5: Compare and Discuss Forecasting Techniques

Determining best-suited forecasting technique
After applying the various forecasting techniques in the previous questions, you are now required to
determine which technique performed the best or is most suitable for your data.

Forecasting Techniques

Compare the various forecasting techniques in the previous 2 questions. Explain which you think is best
suited for the data and why. Include the advantages and disadvantages of each technique, as well as the
characteristics of each, and the outputs from the modeling to justify your answer.

Treat this as an essay-type question and thus structure your answer accordingly.

In analyzing the best-suited forecasting technique for the provided sales data, which consists of monthly sales figures over a period of time, we have explored three main techniques: ARIMA (Autoregressive Integrated Moving Average), SARIMA (Seasonal ARIMA), and ARMA (Autoregressive Moving Average). Each technique has its own advantages, disadvantages, and characteristics, which we'll examine to determine the most suitable approach for forecasting these sales data.

Starting with ARIMA, this technique is well-suited for modeling time series data with trends and/or seasonality. It combines autoregression, differencing, and moving average components to capture both short-term and long-term patterns in the data. In our analysis, we fitted an ARIMA model with an order of (1, 1, 1), indicating one autoregressive term, one differencing term, and one moving average term. This allowed us to capture the autocorrelation and seasonality present in the sales data effectively.

One of the main advantages of ARIMA is its flexibility in modeling various types of time series data, including those with complex patterns. Additionally, ARIMA provides diagnostic tools such as residual analysis and statistical tests to assess the goodness-of-fit of the model. However, ARIMA may struggle with highly irregular or non-stationary data, requiring additional preprocessing steps such as transformation or detrending.

Moving on to SARIMA, this technique extends ARIMA by incorporating seasonal components, making it particularly useful for data with recurring patterns at fixed intervals. By fitting a SARIMA model with a seasonal order of (1, 1, 1, 12), we were able to capture the monthly seasonality present in the sales data. SARIMA accounts for both the autocorrelation within the series and the seasonality, providing a comprehensive framework for forecasting.

The advantage of SARIMA lies in its ability to handle seasonal data without the need for additional preprocessing steps. By explicitly modeling the seasonal component, SARIMA can produce more accurate forecasts for time series with strong seasonal patterns. However, SARIMA may require more parameters to be estimated, leading to increased complexity and potentially overfitting if not properly tuned.

We explored the ARMA technique, which focuses solely on the autoregressive and moving average components without considering seasonality. While ARMA can be effective for modeling stationary time series data with autocorrelation and short-term dependencies, it may not be suitable for datasets with significant seasonal variations, such as the monthly sales data we have.

In conclusion, after assessing the various forecasting techniques applied to the provided sales data, we find that SARIMA is the most suitable approach. SARIMA combines the advantages of ARIMA with the ability to capture seasonal patterns, providing a robust framework for forecasting monthly sales. By explicitly modeling both the non-seasonal and seasonal components, SARIMA produces accurate forecasts while accounting for the underlying patterns in the data. Therefore, SARIMA is recommended for forecasting the provided sales data due to its ability to handle both trend and seasonality effectively, leading to more reliable predictions.
